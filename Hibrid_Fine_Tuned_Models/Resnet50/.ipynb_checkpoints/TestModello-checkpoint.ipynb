```python
from sklearn.metrics import confusion_matrix
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
import os

# Load the pre-trained ResNet50 model
model = load_model("/users/emanueleamato/Downloads/fine_tuned_resnet50_model.h5")

# Define the target image size for ResNet50
target_size = (224, 224)

# Path to the directory containing test images
path = '/users/emanueleamato/Downloads/Immagini_Test'

# Lista delle etichette vere per ogni immagine nel set di test
true_labels = []

# Lista delle previsioni del modello per ogni immagine nel set di test
predicted_labels = []

# Threshold per la classificazione delle immagini
threshold = 0.55

def get_predicted_label(predictions, threshold):
    if predictions[0][0] > threshold:
        return "2CH"
    elif predictions[0][1] > threshold:
        return "4CH"
    else:
        return "Unknown Class"

path = '/users/emanueleamato/Downloads/Test'
# Iterate attraverso ogni immagine nel set di test
for root, dirs, files in os.walk(path):
    for file in files:
        if not file.startswith('.'):
            img_path = os.path.join(root, file)
            true_label = os.path.basename(root)  # Etichetta vera  il nome della cartella padre
            true_labels.append(true_label)

            # Carica e preprocessa l'immagine
            img = image.load_img(img_path, target_size=target_size)
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = preprocess_input(img_array)
            
            # Effettua le previsioni del modello
            predictions = model.predict(img_array)
            predicted_label = get_predicted_label(predictions, threshold)  # Funzione per ottenere l'etichetta prevista
            predicted_labels.append(predicted_label)

# Calcola la matrice di confusione
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Stampa la matrice di confusione
print("Matrice di confusione:")
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt

conf_matrix = np.array(conf_matrix)
conf_matrix = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

# Plot della matrice di confusione
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Matrice di Confusione')
plt.colorbar()
plt.xlabel('Etichetta Prevista')
plt.ylabel('Etichetta Reale')
plt.xticks(np.arange(3), ['2CH', '4CH', 'Unknown'])
plt.yticks(np.arange(3), ['2CH', '4CH', 'Unknown'])

# Mostra i valori all'interno della matrice
thresh = conf_matrix.max() / 2.
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(j, i, format(conf_matrix[i, j], '.2f'),
                 horizontalalignment="center",
                 color="white" if conf_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```
